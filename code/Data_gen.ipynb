```python
# https://competitions.codalab.org/competitions/23601#learn_the_details-instructions
# https://github.com/AnDiChallenge/ANDI_datasets

import andi 
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
from sklearn.preprocessing import MinMaxScaler
# plt.style.use('tsong')


AD = andi.andi_datasets()

model_id = AD.avail_models_name
for i in range(len(model_id)):
    model_id[i] = model_id[i].upper()


# devide due to levy cannot creat sub-diffusion
dim = 1  # 2D, 3D [model_id, exponents, x:tmax, y:tmax, z:tmax]

twmins = [10, 10, 10, 10, 10 ]

for itry, twmin in enumerate (twmins):
    twmax = 1000

    nmax = 10000
    tmax = np.random.randint(low=twmin,high=twmax, size=nmax)
    # tmax = np.random.randint(low=100,high=1000, size=nmax)
    # tmax = np.random.randint(low=10,high=1000, size=nmax)
    sub_exponents_candi = np.random.uniform(low=0.1, high=0.9, size=nmax)
    sup_exponents_candi = np.random.uniform(low=1.1, high=1.9, size=nmax)


    sub_models = [0, 1, 2, 4]
    sup_models = [2, 3, 4]

    dataset = []
    noisy_data = []
    scaled_data = []
    scaled_noisy = []

    labelset = []
    exponset = []

    scaler = MinMaxScaler(feature_range=(0, 1))


    for im, imod in enumerate(sub_models):
        icount = 0
        while icount <= nmax-1:
            trajecs = AD.create_dataset(T = tmax[icount], N = 1, exponents = sub_exponents_candi[icount],\
                                            models = imod, dimension = dim )

            if np.sum(np.square((trajecs[0][2:]))) != 0:
                labelset.append(trajecs[0][0])
                exponset.append(trajecs[0][1])
                dataset.append(trajecs[0][2:])
                icount = icount + 1 


    for im, imod in enumerate(sup_models):
        icount = 0
        while icount <= nmax-1:

            trajecs = AD.create_dataset(T = tmax[icount], N = 1, exponents = sup_exponents_candi[icount],\
                                            models = imod, dimension = dim )

            if np.sum(np.square((trajecs[0][2:]))) != 0:
                labelset.append(trajecs[0][0])
                exponset.append(trajecs[0][1])
                dataset.append(trajecs[0][2:])
                icount = icount + 1




    import pickle
#     f = open('./DATABASE/test_1d_Tmin{0:d}_Tmax{1:d}_N{2:d}.pkl'.format(twmin,twmax,nmax), 'wb')
    # f = open('./DATABASE/long_%id_Tmax1000_N%i.pkl' %(dim,nmax), 'wb')
    # f = open('./DATABASE/task_%id_Tmax1000_N%i.pkl' %(dim,nmax), 'wb')
    f = open('./DATABASE/task_sim_%id_%i_nmax10000.pkl' %(dim,itry), 'wb')

    trans_data = {'trajec': dataset, 'label_id': model_id, 'label': labelset, 'elabel': exponset }

    pickle.dump(trans_data, f)
    f.close()



```
